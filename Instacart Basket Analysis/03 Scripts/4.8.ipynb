## Question 4

### Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# creating a loyalty flag column in your ords_prods_merge dataframe.
```

```python
# all three steps can be included in a single line of code:
# 1. Split the data into groups based on the “user_id” column.
# 2. Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
# 3. Create a new column, “max_order,” into which you’ll place the results of your aggregation.
```

```python
#First, create your “max_order” column.
```

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#check out the results using the head() function:
```

```python
df_ords_prods_merge.head(15)
```

```python
pd.options.display.max_rows = None
```

```python
df_ords_prods_merge.head(100)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

## Question 6

### The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria: 

### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” 

### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
df_ords_prods_merge['average_product_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_product_price']<10, 'customer_income'] = 'Low spender'
```

## Question 7

### In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

### If the median of “days_since_prior_order” is higher than 20, then the customer should  be labeled a “Non-frequent customer.”

### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
df_ords_prods_merge['median_days_since_prior_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check out results using .head() function
```

```python
df_ords_prods_merge.head(15)
```

```python
# A new “median_days_since_prior_order” column has appeared at the end of the dataframe.
# Each value in this column corresponds to the median of the days since the last order made by each user ID
```

```python
#conduct check by printing the head of the dataframe with an argument of 100
```

```python
df_ords_prods_merge.head(100)
```

```python
# create flag
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] > 20, 'frequent_customer'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_since_prior_order'] > 10) & (df_ords_prods_merge ['median_days_since_prior_order'] == 20), 'frequent_customer'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_since_prior_order'] < 10, 'frequent_customer'] = 'Frequent customer'
```

```python
# print the frequency of the new “frequent_customer” column using the value_counts()function:
```

```python
df_ords_prods_merge['frequent_customer'].value_counts(dropna = False) 
```
