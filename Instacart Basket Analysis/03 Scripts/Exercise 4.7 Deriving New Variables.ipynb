## Defining Price Label Command

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# use price_label command
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df.value_counts()
```

```python
df['prices'].max()
```

## If-Statements with the loc() Function

## For-Loop in Instacart

```python
#which day do most orders take place?
# print frequency of the orders_days_of_the_week column
```

```python
df_ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
#in loop to create column "busiest day"
```

```python
result = []

for value in df_ords_prods_merged["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#create new column in data set 
```

```python
df_ords_prods_merged['busiest_day'] = result
```
